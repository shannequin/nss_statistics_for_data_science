1. You randomly select a card from a [standard 52-card deck of playing cards](https://en.wikipedia.org/wiki/Standard_52-card_deck).

```python
print(f"ANSWER: P(Face Card) 12 / 52 = {12/52}")
```

```python
print(f"ANSWER: P(Jack | Diamond) 1 / 13 = {1/13}")
```

2. Say that there is a disease that is carried by 0.1% of the population. A test is developed which gives a positive result 98% of the time when a person has the disease. It also gives a positive result 3% of the time when a person does not have the disease.

	a. What is the probability of a person having the disease and getting a positive result? (Hint: Use the formula for the joint probability which uses the conditional probability)

```python
print(f"ANSWER: P(A and B) = P(A | B) * P(B)")
print(f"P(Diseased and Positive Result) = P(Positive Result | Diseased) * P(Diseased)")
print(f"P(Diseased and Positive Result) = .98 * .001 = {.98 * .001}")
```

	b. What is the probability of a person not having the disease and getting a positive result?

```python
print(f"ANSWER: P(A and B) = P(A | B) * P(B)")
print(f"P(Not Diseased and Positive Result) = P(Positive Result | Not Diseased) * P(Not Diseased)")
print(f"P(Not Diseased and Positive Result) = .03 * .999 = {.03 * .999}")
```

	c. What is the probability of a person getting a positive result? (Hint: Add the probabilities from the previous two questions)

```python
print("ANSWER: P(Diseased and Positive Result) + P(Not Diseased and Positive Result)")
print(f"0.00098 + 0.02997 = {0.00098 + 0.02997}")
```

	d. If you get tested and get a positive result, what is the probability that you have the disease? (Hint: Use Bayes' Theorem)

```python
print("ANSWER: P(A | B) = P(B | A) / P(B) * P(A)")
print("P(Diseased | Positive Result) = P(Positive Result | Diseased) / P(Positive Result) * P(Diseased)")
print(f"P(Diseased | Positive Result) = .98 / .03095 * .001 = {.98 / .03095 * .001}")
```

### Part 2: Independence or Dependence of Variables

1. Let x be the event that it rains today and y be the event that it rained yesterday. Do you think that x and y are independent or dependent?

ANSWER: Dependent

2. Use the data contained in the bna_rain.csv to evaluate your guess. Read this data into a DataFrame named `bna_rain`. The `rained_today` column indicates whether there was any precipitation on a given day and the `rained_yesterday` column indicates whether there was any precipitation the day before. **Hint:** If you want to filter the DataFrame down to the rows for days where it rained yesterday, you can use .loc: `bna_rain.loc[bna_rain['rained_yesterday']]`.

```python
import pandas as pd
import matplotlib.pyplot as plt

bna_rain = pd.read_csv("../data/bna_rain.csv")
```

***P(x|y) = P(x)***
<br>
***P(rained_today|rained_yesterday) = P(rained_today)***

```python
### P(x) ###
rained_today_sum = bna_rain['rained_today'].sum()
print(f"RAINED TODAY SUM: {rained_today_sum}")
total_days = bna_rain['rained_today'].count()
print(f"TOTAL DAYS: {total_days}")
px = rained_today_sum / total_days # Can also use .mean()
print(f"P(x): {px}")

### P(x|y) ###
rained_yesterday_df = bna_rain.loc[bna_rain['rained_yesterday']]
rained_today_sum = rained_yesterday_df['rained_today'].sum()
print(f"RAINED TODAY SUM: {rained_today_sum}")
total_days = rained_yesterday_df['rained_today'].count()
print(f"TOTAL DAYS: {total_days}")
pxy = rained_today_sum / total_days # Can also use .mean()
print(f"P(x|y): {pxy}")
```

***P(x and y) = P(x) * P(y)***
<br>
***P(rained_today and rained_yesterday) = P(rained_today) * P(rained_yesterday)***

```python
### P(y) ###
rained_yesterday_sum = bna_rain['rained_yesterday'].sum()
print(f"RAINED YESTERDAY SUM: {rained_yesterday_sum}")
total_days = bna_rain['rained_yesterday'].count()
print(f"TOTAL DAYS: {total_days}")
py = rained_today_sum / total_days
print(f"P(y): {py}")

### P(x and y) ####
print(f"P(x and y) = {px + py}")
print(f"P(x) * P(y) = {px * py}")
```

1. You take a random sample of 10 American workers.

	a. What is the probability that between 6 and 9 people (inclusive) in your sample are living paycheck-to-paycheck?

$$P(x \leq 9) = P(x \leq 5) + P(6 \leq x \leq 9)$$
<br>
$$P(6 \leq x \leq 9) = P(x \leq 9) - P(x \leq 5)$$

```python
from scipy.stats import binom

print(f"P(6 <= x <= 9): {binom.cdf(k=9, n=10, p=0.78) - binom.cdf(k=5, n=10, p=0.78)}")
```

    b. What is the probability that at least 9 people in your sample are living paycheck-to-paycheck?

$$P(x \geq 9) = P(x > 8) = 1 - P(x \leq 8)$$

```python
print(f"P(x >= 9): {1 - binom.cdf(k=8, n=10, p=0.78)}")
```

	c. What is the probability that 70% or fewer of people in your sample are living paycheck-to-paycheck?

```python
print(f"P(x <= 7): {binom.cdf(k=7, n=10, p=0.78)}")
```

### Part 5: The Normal Distribution

1. Using a standard normal distribution (mean 0, standard deviation 1), answer the following questions.

	a. What percentage of outcomes are within one standard deviation of the mean?

```python
from scipy.stats import norm

# P(-1 < x < 1)
norm.cdf(x=1, loc=0, scale=1) - norm.cdf(x=-1, loc=0, scale=1)
```

	b. What percentage of outcomes are within two standard deviations of the mean?

```python
norm.cdf(x=2, loc=0, scale=1) - norm.cdf(x=-2, loc=0, scale=1)
```

	c. What percentage of outcomes are within three standard deviations of the mean?

```python
norm.cdf(x=3, loc=0, scale=1) - norm.cdf(x=-3, loc=0, scale=1)
```

2. How do your answers change to the previous question if you are using a normal distribution that has a different mean and/or standard deviation?

	f. Challenge Question - If two men are chosen at random, what is the probability that their __mean__ height is between 68 and 72 inches tall? (Hint: You'll probably have to simulate this to get an approximate answer. Similar to what we did to generate binomial random variables, we can generate normal random variables using `norm.rvs()`. For example, `norm.rvs(loc = 70, scale = 3, size = 2)` will generate two observations from a normal random variable with mean 70 and standard deviation 3. You can also give it a tuple.
